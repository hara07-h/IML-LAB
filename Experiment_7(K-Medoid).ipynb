```python
data_points = {'A': 2, 'B': 4, 'C': 5, 'D': 10, 'E': 12}
medoids = [2, 10]
k = 2
clusters = {medoid: [] for medoid in medoids}
def calculate_distance(p1, p2):
    return abs(p1 - p2)
print("Assigning points to closest medoid:")
for label, point in data_points.items():
    distances = {medoid: calculate_distance(point, medoid) for medoid in medoids}
    closest_medoid = min(distances, key=distances.get)
    clusters[closest_medoid].append((label, point))
    print(f"  Point {label}({point}): Distances to medoids: {distances} -> Assigned to medoid {closest_medoid}")
total_cost = 0
print("\nClusters and Costs:")
for medoid, points_in_cluster in clusters.items():
    cluster_cost = 0
    if points_in_cluster:
        print(f"  Medoid {medoid}:")
        for label, point in points_in_cluster:
            dist = calculate_distance(point, medoid)
            cluster_cost += dist
            print(f"    Point {label}({point}), Distance to medoid {medoid}: {dist}")
    total_cost += cluster_cost
    print(f"  Cluster for medoid {medoid} (Points: {[p[0] for p in points_in_cluster]}): Cost = {cluster_cost}")

print(f"\nTotal Cost for this configuration: {total_cost}")

print("\nNext, we would explore swapping a non-medoid point with a medoid to see if the total cost can be reduced.")
```
