```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3]).reshape(-1, 1)
Y = np.array([
    [2, 3],
    [4, 5],
    [6, 7]
])

# Train model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Plot y1
plt.figure(figsize=(6, 4))
plt.scatter(X, Y[:, 0], color='blue', label='Actual y1')
plt.plot(X, Y_pred[:, 0], color='red', label='Predicted y1')
plt.xlabel('X')
plt.ylabel('y1')
plt.title('Multivariate Regression for y1')
plt.legend()
plt.grid(True)
plt.show()

# Plot y2
plt.figure(figsize=(6, 4))
plt.scatter(X, Y[:, 1], color='green', label='Actual y2')
plt.plot(X, Y_pred[:, 1], color='orange', label='Predicted y2')
plt.xlabel('X')
plt.ylabel('y2')
plt.title('Multivariate Regression for y2')
plt.legend()
plt.grid(True)
plt.show()
```
